```python
def load_data(nasa_model_s_data):
    # Skip commented lines and load data into a pandas DataFrame
    data = pd.read_csv(nasa_model_s_data,
        delim_whitespace=True,
        comment='#',
        header=None,
        names=['r/R', 'c (cm/sec)', 'rho (g/cm^3)', 'p (dyn/cm^2)', 'Gamma_1', 'T (K)']
    )
    return data
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_data(filename="/content/nasa_model_s_data.txt"):
    data = pd.read_csv(
        filename,
        delim_whitespace=True,
        comment='#',
        header=None,
        names=['r/R', 'c (cm/sec)', 'rho (g/cm^3)', 'p (dyn/cm^2)', 'Gamma_1', 'T (K)']
    )
    return data
```

```python
def compute_temperature_gradient(data):
    radius = data['r/R']
    temperature = data['T (K)']

    # Numerical derivative using numpy gradient
    dT_dr = np.gradient(temperature, radius)

    return radius, dT_dr
```

```python
def plot_temperature_gradient(radius, dT_dr):
    plt.figure(figsize=(10, 6))
    plt.plot(radius, dT_dr, label=r"$\frac{dT}{dr}$", color='blue')
    plt.xlabel('Radius (r/R)', fontsize=12)
    plt.ylabel(r'Temperature Gradient $\frac{dT}{dr}$ (K per unit r/R)', fontsize=12)
    plt.title('Temperature Gradient vs. Radius', fontsize=14)
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
def main():
    filename = 'nasa_model_s_data.txt'  # Ensure this file contains your data
    data = load_data(filename)

    radius, dT_dr = compute_temperature_gradient(data)
    plot_temperature_gradient(radius, dT_dr)
```

```python
if __name__ == '__main__':
    main()
```

```python
from scipy.optimize import curve_fit
def polynomial_fit(radius, a, b, c, d):
    return a * radius**3 + b * radius**2 + c * radius + d
```

```python
def fit_function(radius, dT_dr):
    params, _ = curve_fit(polynomial_fit, radius, dT_dr)
    return params
```

```python
def plot_fit(radius, dT_dr, params):
    fitted_dT_dr = polynomial_fit(radius, *params)

    plt.figure(figsize=(10, 6))
    plt.scatter(radius, dT_dr, label='Original Data', color='blue')
    plt.plot(radius, fitted_dT_dr, label='Fitted Function', color='red')
    plt.xlabel('Radius (r/R)', fontsize=12)
    plt.ylabel(r'Temperature Gradient $\frac{dT}{dr}$', fontsize=12)
    plt.title('Fitted Function for Temperature Gradient', fontsize=14)
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
def display_function(params):
    print("Fitted Function:")
    print(f"dT/dr = {params[0]:.3e} * r^3 + {params[1]:.3e} * r^2 + {params[2]:.3e} * r + {params[3]:.3e}")
```

```python
def main():
    filename = 'nasa_model_s_data.txt'  # Ensure this file contains your data
    data = load_data(filename)
    radius, dT_dr = compute_temperature_gradient(data)
    params = fit_function(radius, dT_dr)
    display_function(params)
    plot_fit(radius, dT_dr, params)
```

```python
if __name__ == '__main__':
    main()
```

```python
from scipy.integrate import quad
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def load_data(filename):
    data = pd.read_csv(
        filename,
        delim_whitespace=True,
        comment='#',
        header=None,
        names=['r/R', 'c (cm/sec)', 'rho (g/cm^3)', 'p (dyn/cm^2)', 'Gamma_1', 'T (K)']
    )
    return data

def compute_temperature_gradient(data):
    radius = data['r/R']
    temperature = data['T (K)']
    dT_dr = np.gradient(temperature, radius)
    return radius, dT_dr

def polynomial_derivative(radius, a, b, c):
    return 3 * a * radius**2 + 2 * b * radius + c

def polynomial_integral(radius, a, b, c, d):
    return (a / 4) * radius**4 + (b / 3) * radius**3 + (c / 2) * radius**2 + d * radius

def polynomial_fit(radius, a, b, c, d):
    return a * radius**3 + b * radius**2 + c * radius + d

# Step 5: Fit the Polynomial to Data
def fit_function(radius, dT_dr):
    params, _ = curve_fit(polynomial_fit, radius, dT_dr)
    return params

def plot_integrated_function(radius, dT_dr, params):
    # Fitted function and its integral (temperature)
    fitted_temperature = polynomial_integral(radius, *params)

    plt.figure(figsize=(12, 8))
    plt.scatter(radius, dT_dr, label='Numerical Derivative (Original)', color='blue', s=20)
    plt.plot(radius, fitted_temperature, label='Integrated Fitted Function (Temperature)', color='red', linestyle='dashed')
    plt.xlabel('Radius (r/R)', fontsize=12)
    plt.ylabel('Temperature (K)', fontsize=12)
    plt.title('Comparison of Integrated Function vs Original Temperature', fontsize=14)
    plt.grid(True)
    plt.legend()
    plt.show()

def main():
    filename = 'nasa_model_s_data.txt'  # Ensure this file contains your data
    data = load_data(filename)
    radius, dT_dr = compute_temperature_gradient(data)
    params = fit_function(radius, dT_dr)

    # Call plot_integrated_function with the calculated values
    plot_integrated_function(radius, dT_dr, params)

if __name__ == '__main__':
    main()
```
