```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creates an array to represent each education level
Education_levels = np.array([
"Less than high school completion",
"High school completion",
"Some college, no degree",
"Associate’s degree",
"Bachelor’s degree",
"Master’s or higher degree"
])
```

```python
Income_arr = np.array([35500.,41800.,45200.,49500.,66600.,80200.]) # Creates an array that contains the income data
Education_arr = np.arange(len(Income_arr)) # Creates a numerical array to correspond to the education levels
# Plots the income versus education level data
plt.plot(Education_arr,Income_arr, "o-")
# Labels and formats the plot
plt.xlabel("Education Level")
plt.ylabel("Income [$]")
plt.xticks(Education_arr, Education_levels, rotation=45, ha="right")
plt.show()
```

```python
Income_arr_women = np.array([30900.,36700.,39100.,43100.,60800.,75700]) # Creates an array that contains the income data
Education_arr = np.arange(len(Income_arr)) # Creates a numerical array to correspond to the education levels
# Plots the income versus education level data for women specifically
plt.plot(Education_arr,Income_arr, "o-")
# Labels and formats the plot
plt.xlabel("Education Level")
plt.ylabel("Income [$] (Women)")
plt.xticks(Education_arr, Education_levels, rotation=45, ha="right")
plt.show()
```

```python
# Plots each component of the graph (income versus education level in general as well as for women specifically)
plt.plot(Education_arr,Income_arr, color="red", label="Total Income", marker="o", linestyle="-")
plt.plot(Education_arr, Income_arr_women, color="blue", label="Women's Income", marker="o", linestyle="-")
# Labels and formats the plot
plt.xlabel("Education Level")
plt.ylabel("Income [$]")
plt.xticks(Education_arr, Education_levels, rotation=45, ha="right")
plt.legend()
plt.show()
```

```python
z_linear = np.polyfit(Education_arr, Income_arr_women, 1)  # Calculates the linear fit for the women's income data (degree 1)
C_linear = z_linear[0] * Education_arr + z_linear[1] # This gives us the fitted line for the corresponding education levels
O_err = np.array([1910, 500, 2220, 2140, 1540, 2580]) # Gives us an error array for the data
```

```python
# Plots each component of the graph (income versus education level in general as well as for women specifically)
plt.plot(Education_arr,Income_arr, color="red", label="Total Income", marker="o", linestyle="-")
plt.plot(Education_arr, Income_arr_women, color="blue", label="Women's Income", marker="o", linestyle="-")
plt.errorbar(Education_arr, Income_arr_women, yerr=O_err, marker="o", linestyle="None", color="black", label="Data (Women's Income)") # Plots error bars for the women's income versus education level data
plt.plot(Education_arr, C_linear, color="green", label="Linear Fit (Women's Income)") # Plots the linear fit for the women's income versus education level data
# Labels and formats the plot
plt.xlabel("Education Level")
plt.ylabel("Income [$]")
plt.xticks(Education_arr, Education_levels, rotation=45, ha="right")
plt.legend()
plt.show()
```
