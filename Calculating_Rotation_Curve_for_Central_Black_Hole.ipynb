```python
import astropy.constants as ac # import a Python package to call the value of certain constants
import astropy.units as u # import a Python package to call certain units, ex. the mass of the sun
import numpy as np # import a Python package to call certain mathematical functions
```

```python
def calculatingOrbitalVelocity(M, R):
    """
    This is a function that calculates the orbital velocity.
    Input:
    M, the mass of the central object
    R, the orbital radius of the satellite
    Output:
    orbitalVelocity, the orbital velocity of the satellite
    """
    orbitalVelocity = np.sqrt(ac.G * M / R)
    return orbitalVelocity
```

```python
v_galaxyArray = np.arange(1, 31) * u.kpc
print(v_galaxyArray.to(u.m))
```

```python
# Below we define input M and R
M = 1 * 8.54*10**36 * u.kg # mass of Sagittarius A*
R = 1 * v_galaxyArray

# Calculating orbital velocity of the array
v_galaxyArray = calculatingOrbitalVelocity(M, R.to(u.m)) # using the predefined function
print(v_galaxyArray.to(u.km/u.s))
```

```python
import matplotlib.pyplot as plt # importing plotting package

plt.plot(np.arange(1, 31), v_galaxyArray.to(u.km/u.s)) # Plotting velocity vs. orbital radius
plt.xlabel("Orbital Radius (KPC)") # Add x axis label
plt.ylabel("Orbital Velocity (km/s)") # Add y axis label
plt.title("Orbital Velocity vs. Orbital Radius")
plt.show()
```
